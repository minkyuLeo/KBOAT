```python
# value_counts로 'race_id'가 12번 등장하는 값들 찾기
race_ids_with_12 = df_all['race_id'].value_counts()[df_all['race_id'].value_counts() == 12].index

# race_id별로 데이터 분할 후 수정
for race_id in race_ids_with_12:
    # 해당 race_id의 인덱스를 얻기
    indices = df_all[df_all['race_id'] == race_id].index

    # 처음 6개의 데이터는 그대로 두고 뒤의 6개의 데이터는 '_1'을 추가
    df_all.loc[indices[6:], 'race_id'] = df_all.loc[indices[6:], 'race_id'].astype(str) + '_1'

# 결과 출력
print(df_all)
```

```python
# prompt: df_all['race_id'].value_counts()의 값이 6이 아닌것들만 보여줘

value_counts = df_all['race_id'].value_counts()
filtered_value_counts = value_counts[value_counts != 6]
print(filtered_value_counts)
```

```python
# prompt: df_all['race_id'].value_counts()가 6의 값을 가지는 'race_id'를 가진 데이터만 남겨줘

# Find the 'race_id' values with a count of 6
value_counts = df_all['race_id'].value_counts()
race_ids_with_count_6 = value_counts[value_counts == 6].index

# Filter the DataFrame to include only rows with those 'race_id' values
df_filtered = df_all[df_all['race_id'].isin(race_ids_with_count_6)]
df_filtered
```

```python
df_filtered.reset_index(drop=True)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# 예시 데이터프레임 생성 (실제 df_filtered로 대체하세요)
# df_filtered = pd.DataFrame(...)

# 빈 데이터프레임 생성 (결과를 저장할 데이터프레임)
df_standardized = pd.DataFrame()

# 6개 행씩 끊어서 표준화하기
for i in range(0, len(df_filtered), 6):
    # 6개의 행을 선택
    chunk = df_filtered.iloc[i:i+6]

    # 숫자형 데이터만 선택 (비숫자형 데이터가 있는 경우)
    numeric_chunk = chunk.drop(columns=['race_reg_no', 'race_id', 'racer_grd_cd', 'sex_cd','boat_no', 'motor_no','rank123', 'rank12',
       'high_rank'])

    # 스케일러 초기화 및 적용
    scaler = StandardScaler()
    standardized_chunk = scaler.fit_transform(numeric_chunk)

    # 결과를 데이터프레임으로 변환하여 인덱스와 컬럼명 유지
    df_standardized_chunk = pd.DataFrame(standardized_chunk, index=chunk.index, columns=numeric_chunk.columns)

    # 기존의 비숫자형 데이터와 병합 (필요한 경우)
    df_non_numeric = chunk[['race_reg_no', 'race_id', 'racer_grd_cd', 'sex_cd','boat_no', 'motor_no','rank123', 'rank12',
       'high_rank']]
    df_result = pd.concat([df_standardized_chunk, df_non_numeric], axis=1)

    # 결과를 최종 데이터프레임에 추가
    df_standardized = pd.concat([df_standardized, df_result])

# 결과 출력
print(df_standardized)
```
