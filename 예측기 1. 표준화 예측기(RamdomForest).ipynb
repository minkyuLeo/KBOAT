```python
import pandas as pd
```

```python

df = pd.read_csv('/content/24_37_3_표준화.csv')
df
```

```python
df=df.drop(columns=['race_id','boat_no','motor_no'])
```

```python

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, f1_score,precision_score
import joblib  # 모델 저장을 위한 라이브러리
```

```python
best_model = joblib.load('/content/drive/MyDrive/경정데이터공모전/민규/모델/best_random_forest_model.pkl')
model_features=best_model.feature_names_in_ # Access the feature names used during the model training
df = df[model_features]
# 테스트 데이터에 대한 예측
y_pred = best_model.predict(df)
```

```python
# prompt: loaded_model로 예측한 값의 확률을 확인하고싶어

# 예측 확률 얻기
y_pred_proba = best_model.predict_proba(df)

# 예측 확률 출력
#print(y_pred_proba)
df['pred']=y_pred

# 각 클래스에 대한 예측 확률을 DataFrame에 추가
df['pred_proba_0'] = y_pred_proba[:, 0]
df['pred_proba_1'] = y_pred_proba[:, 1]

# 결과 확인
print(df[['pred', 'pred_proba_0', 'pred_proba_1']])
```

```python
df.to_csv('24_37_3_pred.csv',index=False)
```
