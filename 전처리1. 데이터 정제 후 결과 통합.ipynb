```python
# prompt: "mm_6_1_race_high_rank_ratio" ~" mm_6_6_race_high_rank_ratio" 칼럼 각각을 '/' 기준으로 두 칼럼으로 나눠주고, 첫번째칼럼은 mm_6_1_high_rank_ratio, 두번째 칼럼은 mm_6_1_race_count 이런식으로 바꿔줘


# 컬럼 리스트
columns_to_split = ["mm_6_1_race_high_rank_ratio", "mm_6_2_race_high_rank_ratio",
                   "mm_6_3_race_high_rank_ratio", "mm_6_4_race_high_rank_ratio",
                   "mm_6_5_race_high_rank_ratio", "mm_6_6_race_high_rank_ratio"]

# 각 컬럼을 분리하고 새로운 컬럼 생성
for col in columns_to_split:
    df[[col[:-11] + '_rank_ratio', col[:-11] + 'race_count']] = df[col].str.split('/', expand=True)
    df[col[:-11] + '_rank_ratio'] = df[col[:-11] + '_rank_ratio'].astype(float)
    df[col[:-11] + 'race_count'] = df[col[:-11] + 'race_count'].astype(int)

# 기존 컬럼 삭제
# df = df.drop(columns_to_split, axis=1)

df
```

```python
# prompt: stnd_yr,week_tcnt,day_tcnt,race_no는  f"{stnd_yr}_{week_tcnt}_{day_tcnt}_{race_no}" 이런 형태로 합쳐줘, 기존칼럼은 삭제해줘

# 'race_id' 컬럼 생성
df['race_id'] = df.apply(lambda row: f"{row['stnd_yr']}_{row['week_tcnt']}_{row['day_tcnt']}_{row['race_no']}", axis=1)

# 기존 컬럼 삭제
df = df.drop(['stnd_yr', 'week_tcnt', 'day_tcnt', 'race_no'], axis=1)

df
```

```python
# prompt: dptre_tm은 출발시간으로 12:00은 12시00분이라는 뜻인데 시간은 60을 곱하고, 거기다 분을 더해서 하루를 24시간이아니라 1440분의 형태로 나타내고싶어

# 'dptre_tm' 컬럼을 분리하여 시간과 분 추출
df[['hour', 'minute']] = df['dptre_tm'].str.split(':', expand=True).astype(int)

# 1440분 형태로 변환
df['dptre_tm_minutes'] = df['hour'] * 60 + df['minute']

# 기존 'dptre_tm', 'hour', 'minute' 컬럼 삭제
df = df.drop(['dptre_tm', 'hour', 'minute'], axis=1)

df
```

```python
# prompt: row_num,racer_nm,color_nm 은 삭제

df = df.drop(['row_num', 'racer_nm', 'color_nm'], axis=1)
df
```

```python
# prompt: racer_grd_cd는 A1:1, A2 : 2, B1 : 3, B2 : 4  이런식으로 대응되게 바꿔줘, 예를들어 A1은 1로 바꿔줘

# 'racer_grd_cd' 컬럼 값을 숫자로 매핑
grade_mapping = {'A1': 1, 'A2': 2, 'B1': 3, 'B2': 4}
df['racer_grd_cd'] = df['racer_grd_cd'].map(grade_mapping)

df
```

```python
# prompt: sex_cd 성별은 남:0, 여:1 로 바꿔줘

# 'sex_cd' 컬럼 값을 숫자로 매핑
sex_mapping = {'남': 0, '여': 1}
df['sex_cd'] = df['sex_cd'].map(sex_mapping)

df
```

```python
def calculate_average_rank(rank_string):
  # 정수로 변환하기 전에 공백 제거
  ranks = [int(x) for x in rank_string.replace(' ', '') if x != '0']
  if ranks:
    return sum(ranks) / len(ranks)
  else:
    return 0  # 또는 NaN, None 등 적절한 값으로 처리

df['tms_8_rank_ord_no_avg'] = df['tms_8_rank_ord_no'].apply(calculate_average_rank)
df
```

```python
# 'mot_bf_racer_1_no'를 숫자형으로 변환
# 빈 문자열은 NaN으로 변환됩니다.
df_fly['mot_bf_racer_1_no'] = pd.to_numeric(df_fly['mot_bf_racer_1_no'], errors='coerce')

# 각 숫자를 자릿수별로 분리하여 평균 계산
def calculate_average_digit(number):
    # NaN 값 처리
    if pd.isna(number):
        return number

    digits = [int(x) for x in str(int(number)) if x != '0'] # 숫자를 문자열로 변환하기 전에 정수로 변환
    if digits:
        return sum(digits) / len(digits)
    else:
        return np.nan # 일관성을 위해 np.nan 사용

# 새로운 컬럼 생성
df_fly['mot_bf_avg'] = df_fly['mot_bf_racer_1_no'].apply(calculate_average_digit)
df_fly.drop(columns=['mot_bf_racer_1_no'], inplace=True) # DataFrame을 직접 수정하기 위해 inplace=True 사용
df_fly
```

```python
# prompt: 컬럼의 다중공선성 확인

from statsmodels.stats.outliers_influence import variance_inflation_factor

# 독립 변수들만 추출
independent_vars = df
# VIF 계산
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(independent_vars.values, i) for i in range(independent_vars.shape[1])]
vif["features"] = independent_vars.columns
```

```python


# 컬럼별 상관관계 분석
corr = df.corr()

# 컬럼 정렬 (여기서는 VIF 값을 기준으로 정렬)
sorted_columns = vif['features'].tolist()
corr_sorted = corr.loc[sorted_columns, sorted_columns]

# 히트맵으로 시각화
plt.figure(figsize=(30, 20))
sns.heatmap(corr_sorted, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Sorted by VIF)')
plt.show()
```

```python
# prompt: stnd_yr,week_tcnt,day_tcnt,race_no는  f"{stnd_yr}_{week_tcnt}_{day_tcnt}_{race_no}" 이런 형태로 합쳐줘, 기존칼럼은 삭제해줘

# 'race_id' 컬럼 생성
result['race_id'] = result.apply(lambda row: f"{row['stnd_yr']}_{row['week_tcnt']}_{row['day_tcnt']}_{row['race_no']}", axis=1)

# 기존 컬럼 삭제
result = result.drop(['stnd_yr', 'week_tcnt', 'day_tcnt', 'race_no'], axis=1)

# 결과 출력
print(result)
```

```python
import pandas as pd

# 각 컬럼의 첫 번째 글자만 추출하여 숫자로 변환하는 함수
def extract_and_convert(rank):
    if pd.isna(rank):
        return None  # 결측값이 있을 경우 None 반환
    return int(str(rank)[0].replace('①', '1').replace('②', '2').replace('③', '3')
                         .replace('④', '4').replace('⑤', '5').replace('⑥', '6'))

# 새로운 컬럼 생성 (각 rank 컬럼의 첫 글자 모아서 숫자로 변환)
def calculate_rank123(row):
    r1 = extract_and_convert(row['rank1'])
    r2 = extract_and_convert(row['rank2'])
    r3 = extract_and_convert(row['rank3'])

    # 세 값 중 하나라도 None이면 None 반환
    if r1 is None or r2 is None or r3 is None:
        return None
    return r1 * 100 + r2 * 10 + r3

result['rank123'] = result.apply(calculate_rank123, axis=1)
```
