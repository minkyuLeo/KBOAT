### 최종 예측( 123 위 예측 후 1위 예측)

```python

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, f1_score
```

```python
import joblib
model123=joblib.load('/content/drive/MyDrive/경정데이터공모전/민규/모델/123위예측.joblib')
```

```python
model_features=model123.feature_names_in_ # Access the feature names used during the model training
df = df[model_features]
```

```python

X_test=df
```

```python
y_pred_123=model123.predict(X_test)
```

```python
X_test['pred123']= y_pred_123
```

```python
X_test_123 = X_test[X_test['pred123'] == 1]
```

```python
model1of123=joblib.load('/content/drive/MyDrive/경정데이터공모전/민규/모델/123 중에 1위예측.joblib')
```

```python
y_pred_1of123=model1of123.predict(X_test_123.drop(columns=['pred123']))
```

```python
X_test_123['pred1of123']=y_pred_1of123
```

```python
# prompt: df와 df_pred123[['pred1of123']]을 인덱스 기준으로 concat할거고 df에는 있지만 df_pred123[['pred1of123']]에는 없는 경우에는 0으로 대체할거야

import pandas as pd
# df와 df_pred123[['pred1of123']]을 인덱스 기준으로 concat
result_df = pd.concat([X_test, X_test_123[['pred1of123']]], axis=1)

# df에는 있지만 df_pred123[['pred1of123']]에는 없는 경우에는 0으로 대체
result_df['pred1of123'] = result_df['pred1of123'].fillna(0)
```

```python
# 예측 확률 얻기
y_pred_proba = model1of123.predict_proba(X_test_123.drop(columns=['pred123','pred1of123']))

# 예측 확률 출력
print(y_pred_proba)

# 각 클래스에 대한 예측 확률을 DataFrame에 추가
X_test_123['pred_proba_0'] = y_pred_proba[:, 0]
X_test_123['pred_proba_1'] = y_pred_proba[:, 1]

result_df = pd.concat([result_df, X_test_123[['pred_proba_0','pred_proba_1']]], axis=1)
result_df['pred_proba_0'] = result_df['pred_proba_0'].fillna(0)
result_df['pred_proba_1'] = result_df['pred_proba_1'].fillna(0)

# 결과 확인
result_df
```
